### STEP 2: NLP Preprocessing + TF-IDF

```python
import pandas as pd

train = pd.read_csv("/content/merged_data.csv")

df = pd.DataFrame(train)
```

```python
import re

def clean_text(text):
    text = re.sub(r'<.*?>', '', text)   # remove HTML tags
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # keep only letters
    text = text.lower()
    return text

df['review'] = df['review'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

df['review'] = df['review'].apply(
    lambda x: ' '.join([word for word in x.split() if word not in ENGLISH_STOP_WORDS])
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

X = tfidf.fit_transform(df['review']).toarray()
y = df['sentiment']
```

```python
print(X.shape)
```

STEP 3: MODEL TRAINING (2 Models + Comparison)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, batch_train, batch_test = train_test_split(
    X, y, df['batch'], test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

y_pred_nb = nb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def get_metrics(y_true, y_pred):
    return {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred),
        "recall": recall_score(y_true, y_pred),
        "f1": f1_score(y_true, y_pred)
    }

lr_metrics = get_metrics(y_test, y_pred_lr)
nb_metrics = get_metrics(y_test, y_pred_nb)

print("Logistic Regression:", lr_metrics)
print("Naive Bayes:", nb_metrics)
```

```python
results = []

for b in sorted(batch_test.unique()):

    idx = (batch_test == b)

    X_b = X_test[idx]
    y_b = y_test[idx]

    # Logistic Regression
    y_pred_lr_b = lr_model.predict(X_b)

    metrics_lr = get_metrics(y_b, y_pred_lr_b)

    results.append({
        "batch": b,
        "model": "Logistic Regression",
        **metrics_lr
    })

    # Naive Bayes
    y_pred_nb_b = nb_model.predict(X_b)

    metrics_nb = get_metrics(y_b, y_pred_nb_b)

    results.append({
        "batch": b,
        "model": "Naive Bayes",
        **metrics_nb
    })
```

```python
results_df = pd.DataFrame(results)

results_df.head()
```

```python
sentiment_trend = []

for b in sorted(batch_test.unique()):

    idx = (batch_test == b)

    X_b = X_test[idx]

    y_pred = lr_model.predict(X_b)  # best model use kar rahe hain

    positive_ratio = (y_pred == 1).mean()

    sentiment_trend.append({
        "batch": b,
        "positive_ratio": positive_ratio
    })

sentiment_trend_df = pd.DataFrame(sentiment_trend)

sentiment_trend_df.head()
```

```python
df['review_length'] = df['review'].apply(len)

length_drift = df.groupby('batch')['review_length'].mean().reset_index()

length_drift.head()
```

```python
baseline = length_drift['review_length'].iloc[0]

length_drift['drift_score'] = abs(length_drift['review_length'] - baseline) / baseline

length_drift.head()
```

```python
confidence_data = []

for b in sorted(batch_test.unique()):

    idx = (batch_test == b)

    X_b = X_test[idx]

    probs = lr_model.predict_proba(X_b)
    confidence = probs.max(axis=1).mean()

    confidence_data.append({
        "batch": b,
        "avg_confidence": confidence
    })

confidence_df = pd.DataFrame(confidence_data)

confidence_df.head()
```

```python
final_df = results_df.copy()

final_df = final_df.merge(sentiment_trend_df, on="batch")
final_df = final_df.merge(length_drift[['batch','drift_score']], on="batch")
final_df = final_df.merge(confidence_df, on="batch")

final_df.head()
```

```python
final_df.to_csv("monitoring_data.csv", index=False)
```
