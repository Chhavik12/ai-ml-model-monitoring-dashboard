## loaded dataset 
datset was divided into train and split merged it

```python
import pandas as pd

# Load datasets (go one level up using ..)
train = pd.read_csv("../data/raw/train_data.csv")
test = pd.read_csv("../data/raw/test_data.csv")

# Merge both datasets
df = pd.concat([train, test], ignore_index=True)

# Check result
print("Shape:", df.shape)
print(df.head())
```

```python
print(df.columns)
```

renamed column

```python
df = df.rename(columns={
    '0': 'review',
    '1': 'sentiment'
})
```

```python
print(df.columns)
```

data cleaning removed null values and lower all review column data

```python
df = df.dropna()

df['review'] = df['review'].str.lower()
```

```python
import numpy as np

df['batch'] = np.random.randint(1, 11, size=len(df))
```
